# 2. Data Pre-Processing

before we convert anything in the data set, we have to join train & test data set together

Easy to Pre-Processing one time

```python
#Join train & test data set
train['source'] = 'train'
test['source'] = 'test'

data = pd.concat([train,test], ignore_index = True , sort = False)
#Preview data set
print(data.head(10))
```

```python
#Looking for missin value in term of percentage
data.isnull().sum()/data.shape[0]*100
```

You will see Product Category_2 and 3 have missing value 31.38% and 69.65% respectively.

We cannot drop missing value in Category_3 due to it is more than 50% of data so we try to fill missing value 

by adding the new group in Category_2 & 3 by using negative value to represent missing data

```python
#Impute missin value by using negative value for Product_Category_2
data['Product_Category_2'] = data['Product_Category_2'].fillna(-2.0).astype('float')
data['Product_Category_2'].value_counts().sort_index()
```

```python
#Impute missin value by using negative value for Product_Category_3
data['Product_Category_3'] = data['Product_Category_3'].fillna(-2.0).astype('float')
data['Product_Category_3'].value_counts().sort_index()
```

```python
#Removeing Product_Category_1 group 19 & 20 from train set
#Due to in the test set there is no group 19 & 20, not nessary to keep them
condition = data.index[(data.Product_Category_1.isin([19,20]))&(data.source=='train')]
data = data.drop(condition)
```

```python
#Check train data there is no Group 19 & 20
data['Product_Category_1'].value_counts()
```

# 3. Feature Engineering

```python
#Apply function len(unique()) to every data variable
data.apply(lambda x: len(x.unique()))
```

```python
#Turn Gender to binary
gender_dict = {'F':0 , 'M':1}
data['Gender'] = data['Gender'].apply(lambda line: gender_dict[line])
```

```python
data['Gender'].value_counts()
```

```python
#Convert Age to Numerical Values
age_dict = {'0-17': 0, '18-25':1, '26-35':2, '36-45':3, '46-50':4, '51-55':5, '55+':6}
data['Age'] = data['Age'].apply(lambda line: age_dict[line])
data['Age'].value_counts()
```

```python
#Convert City Category to Binary
city_dict = {'A': 0 , 'B':1, 'C':2}
data['City_Category'] = data['City_Category'].apply(lambda line: city_dict[line])
data['City_Category'].value_counts()
```

```python
#Covert Stay_in_Current_City_Years to binary by using dummy variable
Stay = pd.get_dummies(data['Stay_In_Current_City_Years'], drop_first=True)
```

```python
#Concat dummy variable to the data set
data = pd.concat([data,Stay], axis=1)

#Drop some feature that not necessary for prediction target variable
data = data.drop(['User_ID','Product_ID','Stay_In_Current_City_Years'], axis=1)

#Check New Data Set 
print(data.head(10))
```

```python
#Export Data to Data Science Project, save as csv file
#Divide to test set & train set
train = data.loc[data['source']=='train']
test = data.loc[data['source']=='test']

#Drop Unnecessary Column
train.drop('source', axis =1, inplace = True)
test.drop('source', axis=1, inplace = True)

#Export File as Modified Version
train.to_csv('/Users/AdMiN/Desktop/Data Science Projects/train_modified.csv', index=False)
test.to_csv('/Users/AdMiN/Desktop/Data Science Projects/test_modified.csv', index=False)
```

# 4. Modeling

My purpose to use variety models to predict the target varialbe as following

>Multiple Linear Regression

>Ridge Regression

>Lasso Regression

>Decision Tree Regression

>Random Forest Regression

>Gradient Boosting Regression

Each model will perform with base parameter (no tuning) and then take the best model for hyper parameter tuning 

```python
#Import Modified Data Set
col_names = ['gender', 'age', 'occ', 'city', 'marital','pc1','pc2','pc3','purchase','stayyr_1','stayyr_2', 'stayyr_3','stayyr_4+']
train_modified = pd.read_csv('/Users/AdMiN/Desktop/Data Science Projects/Black Friday Practice Problem/black friday train_modified.csv', header=0, names=col_names)
test_modified = pd.read_csv('/Users/AdMiN/Desktop/Data Science Projects/Black Friday Practice Problem/black friday test_modified.csv', header=0, names=col_names)
```

```python
print(train_modified.head())
```

# Split Data to Train & Test Set From Train_Modified

```python
#Import important model
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error as MSE

#Define Features and target variable from train_modified
X = train_modified.drop('purchase',axis=1) #Features
y = train_modified['purchase'] #target variable

#Split data to train set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# 1. Linear Regression Model

```python
#Import Linear Regression Model
from sklearn.linear_model import LinearRegression

#Build Linear Regression Model
lr = LinearRegression(normalize=True)

#Cross Validation with 5 fold
cv_lr = -cross_val_score(lr, X_train, y_train, scoring='neg_mean_squared_error', cv=5)
```

```python
#Linear Regression Performance from R-Square
mean_MSE = np.mean(cv_lr)
RMSE = np.sqrt(mean_MSE)
print('RMSE of Linear Regression:{:.2f}'.format(RMSE))
```

# 2. Ridge Regression Model

```python
#import Ridge Regression Model
from sklearn.linear_model import Ridge

#Build Ridge Regression Model and set alpha = 0.1
ridge = Ridge(alpha=0.1, normalize = True)

#Cross Validation with 5 fold
cv_ridge = -cross_val_score(ridge, X_train, y_train, scoring='neg_mean_squared_error',cv=5)
```

```python
#Ridge Regression Performance from RMSE
mean_MSE = np.mean(cv_ridge)
RMSE = np.sqrt(mean_MSE)
print('RMSE of Linear Regression:{:.2f}'.format(RMSE))
```

# 3. Lasso Regression Model

```python
#import Lasso Regression Model
from sklearn.linear_model import Lasso

#Build Lasso Regression Model and set alpha = 0.1
lasso = Lasso(alpha=0.1, normalize = True)

#Cross Validation with 5 fold
cv_lasso = -cross_val_score(lasso, X_train, y_train, scoring='neg_mean_squared_error',cv=5)
```

```python
#Lasso Regression Performance from RMSE
mean_MSE = np.mean(cv_lasso)
RMSE = np.sqrt(mean_MSE)
print('RMSE of Linear Regression:{:.2f}'.format(RMSE))
```

# 4. Decision Tree Regression

```python
#import important Decision Tree Regression
from sklearn.tree import DecisionTreeRegressor

SEED = 123

#Build Decision Tree Regression
dt = DecisionTreeRegressor(random_state=SEED)

#Cross Validation with 5 fold
cv_dt = -cross_val_score(dt, X_train, y_train, scoring='neg_mean_squared_error',cv=5)
```

```python
#DecisionTree Regression Performance from RMSE
mean_MSE = np.mean(cv_dt)
RMSE = np.sqrt(mean_MSE)
print('RMSE of Linear Regression:{:.2f}'.format(RMSE))
```

# 5. Random Forest

```python
#import important Random Forest Regression
from sklearn.ensemble import RandomForestRegressor

SEED = 123

#Build Random Forest Regressrion with n_estimators = 100
rf = RandomForestRegressor(n_estimators=100,random_state=SEED)

#Cross Validation with 5 fold
cv_rf = -cross_val_score(rf, X_train, y_train, scoring='neg_mean_squared_error',cv=5)
```

```python
#Random Forest Regression Performance from RMSE
mean_MSE = np.mean(cv_rf)
RMSE = np.sqrt(mean_MSE)
print('RMSE of Linear Regression:{:.2f}'.format(RMSE))
```

```python
#Fit Random Forest Model to Training Set
rf.fit(X_train, y_train)

#Feature Contribution
feature_cols = ['gender', 'age', 'occ', 'city', 'marital','pc1','pc2','pc3','stayyr_1','stayyr_2', 'stayyr_3','stayyr_4+']
feature_imp = pd.Series(rf.feature_importances_, index = feature_cols).sort_values(ascending = False)
feature_imp

#Create Bar Plot
plt.figure(figsize=(10,5))
sns.barplot(x=feature_imp, y=feature_imp.index)
plt.xlabel('Feature Importance Score')
plt.ylabel('Feature')
plt.title('Visualize Important Features by Random Forest Regression')
plt.show()
```

# 6. Gradient Boosting

```python
#import important Gradient Boosting
from sklearn.ensemble import GradientBoostingRegressor

SEED = 123

#Build Gradient Boosting Regression with n_estimator =100, max_depth =3
gb = GradientBoostingRegressor(n_estimators=100, random_state=SEED)

#Cross Validation with 5 fold
cv_gb = -cross_val_score(gb, X_train, y_train, scoring='neg_mean_squared_error', cv=5)
```

```python
#Gradient Boosting Regression Performance from RMSE
mean_MSE = np.mean(cv_gb)
RMSE = np.sqrt(mean_MSE)
print('RMSE of Gradient Boosting Regression:{:.2f}'.format(RMSE))
```

# Conclusion

The best model that give lowest RMSE is Gradient Boosting. All model we just use default paramer as base parameter to measure performance. Next we gonna take Gradient Boosting to Hyper Parameter Tuning to find the optimal parameter

```python
gb.get_params()
```

# Hyper Parameter Tuning of Gradient Boosting

```python
#Fit GB Model to Training Set
gb.fit(X_train, y_train)

#Feature Contribution
feature_cols = ['gender', 'age', 'occ', 'city', 'marital','pc1','pc2','pc3','stayyr_1','stayyr_2', 'stayyr_3','stayyr_4+']
feature_imp = pd.Series(gb.feature_importances_, index = feature_cols).sort_values(ascending = False)
feature_imp

#Create Bar Plot
plt.figure(figsize=(10,5))
sns.barplot(x=feature_imp, y=feature_imp.index)
plt.xlabel('Feature Importance Score')
plt.ylabel('Feature')
plt.title('Visualize Important Features by Gradient Boosting Regression')
plt.show()
```

Parameters for tuning are n_estimators, max_features, max_depth, min_samples_split, min_samples_leaf,learning_rate

```python
#Import important library
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import cross_val_score

#Set parammeters for grid search
#Learning Rate Set
learning_rate = [0.01, 0.05, 0.1]

#Max Features Set
max_features = ['auto', 'sqrt']

#n_estimators Set
n_estimators = [200, 500, 1000]

#Max Depth Set
max_depth = np.arange(5,21,5)

#Parameter Random Search 
params_grid = {'n_estimators':n_estimators,
               'max_depth':max_depth,
               'max_features':max_features,
               'learning_rate':learning_rate    
               }
```

```python
#Build Gradient Boosting Regression Model
gb = GradientBoostingRegressor(random_state = SEED)

#Grid Search CV
gb_cv = GridSearchCV(gb, params_grid, verbose=2, cv=5, n_jobs=-1 )
```

```python
#Fit Random Search CV to Training Set
gb_cv.fit(X_train, y_train)
```
